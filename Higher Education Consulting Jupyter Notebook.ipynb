## **<u>Engineering features: Creating new columns to derive additional analysis from:**</u>




#### Creating a column called "instate_totalcost", which adds together all the costs for someone attending a school living in the same state:


```python
df['instate_totalcost'] = df['in-state tuition'] + df['room'] + df['board'] + df['add. fees'] + df['estim. book costs'] + df['estim. personal $']
```

#### Creating a column called 'outofstate_total_cost', which adds together all the costs for attending a school within their own state

```python
df['outofstate_totalcost'] = df['out-of-state tuition'] + df['room'] + df['board'] + df['add. fees'] + df['estim. book costs'] + df['estim. personal $']
```

#### Creating a column called 'outofstate_difference', which figures out the difference in cost between out of state tuition and instate tuition:

```python
df['outofstate_difference'] = df['out-of-state tuition'] - df['in-state tuition']
```

#### Creating a column called 'percent accepted', which calculates the percentage of the number of applications accepted from the number of applications recieved:

```python
df['percent accepted'] = df['# appl. accepted'] / df["# appli. rec'd"]
```

#### Creating a column called 'percent new enrolled', which calculates the percentage of the number of newly enrolled students based off of the number of applications accepted:

```python
df['percent enrolled'] = df['# new stud. enrolled'] / df['# appl. accepted']
```

#### Creating a column called 'total undergrads', which figures out the total number of undergraduate students from the numbers of full time and part time columns

```python
df['total undergrads'] = df['# FT undergrad'] + df['# PT undergrad']
```

### **Visualization #1: Countplot of the number of colleges per US state**

```python
sns.set(style = 'darkgrid', font_scale = 1)
plt.figure(figsize = (20,11))
sns.countplot(data = df,
              x = 'State',
              order = df['State'].value_counts().index,
              hue = 'Public (1)/ Private (2)',
             palette = ['red', 'blue']);
plt.title("Distribution of Public and Private Schools per State")
plt.xlabel("State")
plt.ylabel("Number of Schools")
plt.savefig('Countplot-numschoolsperstate');
```
